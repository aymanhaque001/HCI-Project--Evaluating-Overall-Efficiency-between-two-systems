```python
"""Necessary Functions"""


"""
av_error_time_per_task(sys): 

This function calculates avg time and error per task for any given system with any
number of tasks. Returns a class with avg error and time. 

"""
def av_error_time_per_task(sys):
    
    av_err ={}
    av_time ={}
    
    for x in sys.errors.columns:
        v=sys.errors[x].mean()
        av_err.update({"Task {}".format(x) : v})
        
    for x in sys.time.columns:
        c=sys.time[x].mean()
        av_time.update({"Task {}".format(x) : c})
   
    class System_avg:
        error = av_err
        time = av_time
        
    return System_avg

"""
Usability_equation_task (sys_av):

    Calculates the Usability per task of a system from the class returned by av_error_time_per_task
    
    using the formula:

    Usability = 1000/(sys_av_error + sys_av_time)
"""        
def Usability_equation_task (sys_av) :
    efficiency = {}
    for i in sys_av.error:
        usability= 1000/(sys_av.error[i] + sys_av.time[i])
        efficiency.update({i : usability})
    return efficiency


""" 

Usability_per_task(sys1,sys2):

takes in two systems. Uses function usability_equation_task to find the udability values per task 
asnd store in. a dictionary. Prints the dictionary using print_dict and also the best system per task with the 
help of function print_best_per_task

"""

def Usability_per_task(sys1,sys2):
    
    usability_per_task= {}
    
    sys1_avg = av_error_time_per_task(sys1) 
    sys2_avg = av_error_time_per_task(sys2)
    sys1_usability = Usability_equation_task(sys1_avg)
    sys2_usability= Usability_equation_task(sys2_avg)
    
    if sys1.__name__!=sys2.__name__ :
        usability_per_task.update({sys1.__name__ : sys1_usability})
        usability_per_task.update({sys2.__name__ : sys2_usability})
    else:
        usability_per_task.update({sys1.__name__ : sys1_usability})
        usability_per_task.update({sys2.__name__+ "(2)" : sys2_usability})

    print("****** The Usability of the Systems per Task ******\n\n")
    
    print_dict(usability_per_task)
    
    print_best_per_task(usability_per_task)
#     return usability_per_task
    
def Usability_equation_sys(time_av, err_av):
    usability_sys = (10**4)/(time_av + err_av)
    return usability_sys

"""
Evaluate_System(Sys1,Sys2):

Takes in the two system classes as parameters. Uses the error and time per user to calculate averages. Uses
Usability_equation_sys to calculate system usabilities with the average values. prints the usability per system and
the best system with function print_usability_per_sys. 

"""

def Evaluate_System(Sys1, Sys2): 
    
    Usability={}
    
    sys1_time_av=Sys1.time.sum(axis=1).mean()
    sys1_error_av=Sys1.errors.sum(axis=1).mean()
    sys2_time_av=Sys2.time.sum(axis=1).mean()
    sys2_error_av=Sys2.errors.sum(axis=1).mean()
    
    if Sys1.__name__!= Sys2.__name__ :
        Usability.update({Sys1.__name__ : Usability_equation_sys(sys1_time_av,sys1_error_av)})
        Usability.update({Sys2.__name__ : Usability_equation_sys(sys2_time_av,sys2_error_av)})
    else:
        Usability.update({Sys1.__name__ : Usability_equation_sys(sys1_time_av,sys1_error_av)})
        Usability.update({Sys2.__name__ +"(2)": Usability_equation_sys(sys2_time_av,sys2_error_av)})
        
    print_usability_per_sys(Usability)
    
    #return Usability

def print_dict(c):
    for i in c.keys():
        print(i)
        print("\n")
        for v in c[i].keys():
            print(v," :", c[i][v])
        print("\n")
        
    print("\n\n")
    
def print_best_per_task(c):
    name=[]
    for i in c.keys():
        name.append(i)
    
    comp={}
    for i in c[name[0]]:
        if c[name[0]][i] > c[name[1]][i]:
            comp.update({i:(name[0])})
        elif c[name[0]][i] == c[name[1]][i]:
            comp.update({i:"Both"})
        else:
            comp.update({i:(name[1])})

    print("\n ****** Best System per task ****** \n\n")
    for key in comp:
        print("for ",key,"is ", comp[key]," || using evaluation heuristic, Usability= 1000/(average error per task + average time per task) ","\n")

        
def print_usability_per_sys(c):

    print("\n\n****** Usability Score per System ****** \n\n")
    name =[]
    for  i in c.keys():
        name.append(i)
        print("For ",i,", Usability Score is: ",c[i],"\n") 
    print("\n \n****** SYSTEM EVALUATION ****** \n")   
    if c[name[0]] > c[name[1]]:
        print("\n ",name[0]," has a better Usability score. Hence ", name[0]," is the better system")
    
    elif c[name[0]] == c[name[1]]:
        print("\n  Both systems have equal usability score, hence both systems are equally good.\n")
    
        
    
    else:
        print("\n ",name[1]," has a better Usability score. Hence ", name[1]," is the better system *****\n")
    
    print("\n \n Evaluation Heuristic used: \n \n System Usability = (1 x 10^4)/(Average of errors by all users + Average of time taken by all users)")
```
